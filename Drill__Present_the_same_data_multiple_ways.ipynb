```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```
happy = pd.read_csv("https://raw.githubusercontent.com/RRamirez21/ThinkfulDrills/master/2017.csv")
```

```
happy.columns
```

```
happy.describe()
```

# One Variable, Four Plots

```
sns.set(style="whitegrid")

g = sns.factorplot(y="Happiness.Score", data=happy,
                   size=6, kind="point", palette="pastel", ci=95)
g.despine(left=True)
plt.title('Happiness Score Pointplot')
plt.show()
```

The pointplot shows an estimate of central tendency, but is not descriptive beyond that.

```
sns.set(style="whitegrid")

ax = sns.boxplot(y='Happiness.Score', data=happy, palette='pastel')  

plt.title('Happiness Score Boxplot')

sns.despine(offset=6, trim=True)

plt.show()
```

This boxplot shows us more information, including the quartiles so we have a better idea of the distribution of values, along with the range.

```
sns.set(style="darkgrid")

sns.distplot(happy["Happiness.Score"])
```

This histogram gives us a more accurate depiction of the distribution of the values. We can also see if there is any skew or kurtosis.

```
sns.set(style="whitegrid")

g = sns.factorplot(y="Happiness.Score", data=happy,
                   size=6, kind="violin", palette="pastel", ci=95)
g.despine(left=True)
plt.title('Happiness Score Violinplot')
plt.show()
```

Lastly, this violinplot is similar to the histogram because it shows the values laid out horizontally. However, it also includes the boxplot inside so we can see the quartiles.

# Two Continuous Variables, Three Plots

```
sns.jointplot(x=happy["Trust..Government.Corruption."], y=happy["Family"], kind='hex')

```

This hex map is good at showing us the values of each category and where they relate. It also includes a p-value to indicate the correlation between the two categories.

```
sns.jointplot(x=happy["Trust..Government.Corruption."], y=happy["Family"], kind='kde')
```

This contour map is similar to the hex map above, but shows us a more detailed area of our most frequently occuring values.

```
g = sns.lmplot(y='Family', 
               x='Trust..Government.Corruption.',
               data=happy,
               fit_reg=True, 
               scatter_kws={'alpha':0.4})
g.set_ylabels("Perception of Corruption in relation to Happiness Score")
g.set_xlabels("Family in relation to Happiness Score")
plt.title('Scatterplot with regression line: Extent to which Perception of Corruption and Family contribute to Happiness Score')
plt.show()
```

This scatterplot shows us the distribution of values and also has a regression line to show the best fit between the categories.

# One Continuous Variable, One Categorical Variable, Six Plots

```
sns.factorplot(x="Generosity", y="Country", kind="bar", size=18, data=happy);
plt.title('Generosity Score by Country Barplot')
plt.show()
```

This barplot is a good way to show us the value of "Generosity" for each country. However, the large amount of countries makes this difficult to easily read and compare.

```
plt.figure(figsize=(20,25))
plt.stem(happy['Country'])
plt.xticks(happy['Generosity'])
plt.gca().invert_yaxis()
plt.title("Generosity Score by Country Lolliplot")


plt.show()
```

This lolliplot is similar to the barplot above except it also has a point to indicate the generosity score for each country. However, because this does not show the actual values, it's only a relative view of the score therefore not as useful.

```
f, axes = plt.subplots(1, 4)
sns.boxplot(  y="Generosity", data=happy,  orient='v' , ax=axes[0])
sns.boxplot(  y="Economy..GDP.per.Capita.", data=happy,  orient='v' , ax=axes[1])
sns.boxplot(  y="Health..Life.Expectancy.", data=happy,  orient='v' , ax=axes[2])
sns.boxplot(  y="Freedom", data=happy,  orient='v' , ax=axes[3])
sns.plt.show()
```

Since each country only has one value for any of these categories, a boxplot of each country would be useless. However, by comparing the boxplots of several categories, we can see that Health/Life Expectancy and Freedom are the scores that are more closely related with a country's happiness score.

```
sns.factorplot(x="Generosity", y="Country", kind="point", size=15, data=happy);
plt.title('Generosity Score by Country Pointplot')
plt.show()
```

This pointplot shows us the generosity score for each country. The line is helpful in this case only because we can follow the path to see the score of each country from top to bottom or vice-versa. However, it would be more useful in recognizing a pattern. For example if we were to sort countries by region, we could see how much the generosity scores vary.

```
sns.set(style="darkgrid")

g = sns.factorplot(x="Generosity", y="Country", data=happy,
                   size=15, kind="point", palette="muted", ci=95)
g.despine(left=True)
plt.title('Generosity Score by Country Scatterplot')
plt.show()
```

This scatterplot is similar to the barplot in that it shows each county's "Generosity" value. However, it is slightly more legible since the barplot takes up more space creating a more cluttered appearance. The only difficulty can be in mistakenly reading values that are close to each other.

```
plt.figure(figsize=(10,25))

plt.scatter(happy["Generosity"], happy["Country"], s=happy["Generosity"]*1000, c='red', alpha=0.5)
plt.title('Generosity Score by Country Bubbleplot')

plt.show()
```

This Bubbleplot is similar to the scatterplot above. However, countries with a higher generosity score appear with a larger bubble so they are easier to distinguish.
