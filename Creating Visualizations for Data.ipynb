```
import seaborn as sns

import numpy as np

df = df.dropna(subset=['FTHG'])
sns.distplot(df['FTHG'], kde=False)
```

This distribution shows the majority of Full Time Home Goals are scored between 0 and 2. 1 goal appears to be the most common and there are a few outliers toward the highger scoring end.

```
sns.set(style="whitegrid")

ax = sns.boxplot(x = df['FTHG'])
```

Just like the curve above, this boxplot indicates most of Full Time Home Goals are between 1 and 2 goals. This aligns with the mean of 1.52 pulled from  the description above. Again, we see outliers from the dots beginning at 4 goals and ending at 9 goals.

```
sns.set(style="whitegrid")

ax = sns.boxplot(data=df)
```

Here we see a larger boxplot which includes the goals for both Home and Away teams at Half Time and Full Time. Both Home and Away boxplots are relatively similar. However, we see that the Full Time Away Goals has a 25% quartile that begins at 0, which is lower than the Full Time Home Goals.

```
import seaborn as sns; sns.set()
import matplotlib.pyplot as plt
ax = sns.lmplot(x="FTHG", y="HTHG", data=df)
```

Here we have a basic scatter plot that shows our Half Time Home Goals compared to the Full Time Home Goals. It appears to somewhat follow our line as we expect that scores at the end of Half Time will generally mean more scores by the end of Full Time.

```
ax = sns.violinplot(x="FTAG", y="FTHG", data=df, scale="count")
```

Here is a violin plot that shows our Full Time Home Goals compared to the Full Time Away Goals. The violin curves appear relatively symmetrical as both Home and Away goals have little variance between them. Additionally, we see that games are typically low scoring based on the scale of the intial violin plots.
